```python
# Fetch real time data from OpenAQ API

import requests
import pandas as pd
import json

# Define a function
def fetch_openaq_durgapur(days_back=90):
    """Fetch recent PM2.5 and PM10 readings for Durgapur from OpenAQ."""
    
    url = "https://api.openaq.io/v2/locations"
    params = {
        "city": "durgapur",
        "country": "IN",
        "limit": 10
    }

    response = requests.get(url, params=params)
    data = response.json()
    
    # Extract location IDs
    locations = data.get("results", [])
    print(f"Found {len(locations)} monitoring locations in Durgapur")
    
    for loc in locations:
        print(f"  - {loc['name']} | ID: {loc['id']} | Sensors: {loc.get('sensors', [])}")
    
    return locations
```
